```python
X = df1.drop('PS_COVID_Faelle',axis=1).values
y = df1['OG_PI_COVID_Faelle'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.8,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 12
generator = TimeseriesGenerator(X_train,X_train, length=length, batch_size=1)
```

```python
X,y = generator[0]
```

```python
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,Activation,Dropout,LSTM
from tensorflow.keras.optimizers import Adam
```

```python
#bitte 3 werte Droppen
n_features = 3
```

```python
length=12
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(length, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
validation_generator = TimeseriesGenerator(X_test,X_test, length=length, batch_size=1)
```

```python
model.fit_generator(generator,epochs=20,
                    validation_data=validation_generator,
                   callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df1)
```

```python
length = 12 # Length of the output sequences (in number of timesteps)
n_features=4
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length=length, batch_size=1)
```

```python
from tensorflow import keras
opt = keras.optimizers.Adam(learning_rate=0.0001)
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(length, n_features)))
model.add(Dense(1))

model.compile(optimizer=opt, loss='mse')


# fit model
model.fit_generator(generator,epochs=10)
```

```python
forecast = []
# Replace periods with whatever forecast length you want
periods = 3

first_eval_batch = scaled_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(periods):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    forecast.append(current_pred) 
    
    # update batch to now include prediction and drop first value
#     current_batch = np.append(current_batch[:,:1,:],[[current_pred]],axis=1)
```
